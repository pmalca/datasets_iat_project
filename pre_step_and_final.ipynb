```python
import regex as re 

# # Define patterns
pattern1 = r'J\s*U\s*E\s*Z\s*:\s*'
pattern3 = r'\n'

# Extract text using regex
def extract_text(row):
    opinion_text = row['opinion_text']
    
    # Ensure the opinion_text is a string
    if not isinstance(opinion_text, str):
        return None
    
    match1 = re.search(pattern1, opinion_text, flags=re.IGNORECASE | re.DOTALL)

    if match1:
        start_index = match1.end()
        # Find the next newline after the start_index
        match3 = re.search(pattern3, opinion_text[start_index:], flags=re.IGNORECASE | re.DOTALL)
        
        if match3:
            end_index = start_index + match3.start()
        else:
            end_index = len(opinion_text)

        return opinion_text[start_index:end_index].strip()
    else:
        return None

documents_amag['judge_from_opinion_7'] = documents_amag.apply(extract_text, axis=1)
documents_amag['judge_from_opinion_7']
```

## Working with 50 samples

```python
unique_values = df_sample['expediente_num'].unique()[:50]  

filtered_dataframes = {}

for value in unique_values:
    filtered_df = df_sample[df_sample['expediente_num'] == value]
    
    filtered_df = filtered_df.sort_values(by='num')
    
    filtered_dataframes[value] = filtered_df

filtered_dataframes
```

```python
# Function to determine decision based on the third half of the text
def determine_decision(text):
    len_text = len(text)
    third_last_part_start = 2 * len_text // 3
    third_last_part = text[third_last_part_start:]

    if any(keyword.lower() in third_last_part.lower() for keyword in keywords):
        return 'Possible F_Decision'
    else:
        return 'Pre-step Decision'

keywords = ['se resuelve:', 'decide:', 'falla:', 'fallo:',
            'archivar el presente proceso', 'rechazándose la demanda', 
            'archívese', 'archivese', 'reformandola', 'fundada en parte la demanda', 'infundada la demanda', 
            'fundada la demanda', 'improcedente la demanda']

for value, df in filtered_dataframes.items():
    df['Decision'] = df['text'].apply(determine_decision)
    
    
```

```python
count_possible_f_decisions = sum(df['Decision'].eq('Possible F_Decision').sum() for df in filtered_dataframes.values())

print(f'Number of "Possible F_Decisions" across all DataFrames: {count_possible_f_decisions}')
```
