```python
import pandas as pd
```

```python
df_2018 = pd.read_csv(r'D:\Proyectos\amag\classification\intermediate\judgmments_outcomes_2018.csv')
df_2019 = pd.read_csv(r'D:\Proyectos\amag\classification\intermediate\judgmments_outcomes_2019.csv')
df_2020 = pd.read_csv(r'D:\Proyectos\amag\classification\intermediate\judgmments_outcomes_2020.csv')
df_2021 = pd.read_csv(r'D:\Proyectos\amag\classification\intermediate\judgmments_outcomes_2021.csv')
df_2022 = pd.read_csv(r'D:\Proyectos\amag\classification\intermediate\judgmments_outcomes_2022.csv')

df = pd.concat([df_2018,df_2019,df_2020,df_2021,df_2022], ignore_index=True)
```

```python
df['outcome'] = df['outcome'].str.replace('Partial', 'PARCIAL').str.replace('Negative', 'DESFAVORABLE').str.replace('Positive', 'FAVORABLE').str.replace('Ambiguous', 'AMBIGUO')
```

```python
df
```

```python
import matplotlib.pyplot as plt

df_2 = pd.read_excel(r"D:\Proyectos\amag\judicial_decisions_classif.xlsx")

merged_df = pd.merge(df, df_2, left_on='case_file', right_on='syvs_id', how='inner')
merged_df['is_match'] = merged_df['outcome'] == merged_df['gpt3_answer']
merged_df
```

```python
match_counts = merged_df['is_match'].value_counts()

# Plot the is_match in a pie chart
labels = ['Match', 'Mismatch']
colors = ['#66b3ff', '#ff9999']
plt.pie(match_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Comparison of Regex and GPT-3 Answers')
plt.show()
```
