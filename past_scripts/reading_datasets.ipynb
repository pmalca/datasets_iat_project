```python
import pandas as pd
from fuzzywuzzy import fuzz

# Function to check for similarity
def check_similarity(value, comparison_list, threshold=90):
    for comparison_value in comparison_list:
        if fuzz.ratio(value, comparison_value) >= threshold:
            return True
    return False

# Get the list of values to compare against
comparison_list_apellido_nombre = juez_df['participant_apellido_nombre'].tolist()
comparison_list_nombre_apellido = juez_df['participant_nombre_apellido'].tolist()

# Check for similarities
similarities = []
for value in jueces['0']:
    if (check_similarity(value, comparison_list_apellido_nombre) or 
        check_similarity(value, comparison_list_nombre_apellido)):
        similarities.append(value)

# Print the matching observations
print("Matching Observations:")
print(similarities)

# Optionally, save the matching observations to a new file
matching_df = jueces[jueces['0'].isin(similarities)]
matching_df
```

```python
def check_similarity(value, comparison_list, threshold=90):
    for comparison_value in comparison_list:
        if fuzz.ratio(value, comparison_value) >= threshold:
            return comparison_value
    return None

# Get the list of values to compare against
comparison_list_apellido_nombre = juez_df['participant_apellido_nombre'].tolist()
comparison_list_nombre_apellido = juez_df['participant_nombre_apellido'].tolist()

# Check for similarities and track the matching value
similarities = []
matching_values = []

for value in jueces['0']:
    match_apellido_nombre = check_similarity(value, comparison_list_apellido_nombre)
    match_nombre_apellido = check_similarity(value, comparison_list_nombre_apellido)
    
    if match_apellido_nombre:
        similarities.append(value)
        matching_values.append(match_apellido_nombre)
    elif match_nombre_apellido:
        similarities.append(value)
        matching_values.append(match_nombre_apellido)

# Create a new DataFrame with matching observations and the corresponding match
matching_df = jueces[jueces['0'].isin(similarities)].copy()
matching_df['amag_ii'] = matching_values
matching_df.rename(columns={'0': 'documents_amg'}, inplace=True)

matching_df
```
