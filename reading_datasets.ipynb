```python
jueces = pd.read_csv(r"D:\Proyectos\amag\data_cleaned_test\temp\judges_2018.csv")
jueces
```

```python
import pandas as pd
from fuzzywuzzy import fuzz

# Function to check for similarity
def check_similarity(value, comparison_list, threshold=90):
    for comparison_value in comparison_list:
        if fuzz.ratio(value, comparison_value) >= threshold:
            return True
    return False

# Get the list of values to compare against
comparison_list_apellido_nombre = juez_df['participant_apellido_nombre'].tolist()
comparison_list_nombre_apellido = juez_df['participant_nombre_apellido'].tolist()

# Check for similarities
similarities = []
for value in jueces['0']:
    if (check_similarity(value, comparison_list_apellido_nombre) or 
        check_similarity(value, comparison_list_nombre_apellido)):
        similarities.append(value)

# Print the matching observations
print("Matching Observations:")
print(similarities)

# Optionally, save the matching observations to a new file
matching_df = jueces[jueces['0'].isin(similarities)]
matching_df
```

```python
unique_values_count = matching_df['0'].nunique()

# Print the result
print(f"The number of unique values in column '0' is: {unique_values_count}")
```
